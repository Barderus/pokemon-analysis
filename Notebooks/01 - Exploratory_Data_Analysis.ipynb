```python
df_diff["capture_rate"] = pd.to_numeric(df_diff["capture_rate"], errors="coerce")
```

```python
df_diff["capture_rate"].dtype
```

```python
numeric_feats = ["capture_rate", "base_egg_steps", "experience_growth"]

# Z-score standardization
for col in numeric_feats:
    df_diff[col + "_zscore"] = stats.zscore(df_diff[col])
    print(df_diff[col + "_zscore"])

df_diff.columns
```

```python
# Invert capture rate (so higher = harder)
df_diff["capture_rate_hard_zscore"] = -df_diff["capture_rate_zscore"]
```

```python
# Build difficulty score
score_components = ["capture_rate_hard_zscore", "base_egg_steps_zscore"]
```

```python
df_diff["difficulty_score"] = df_diff[score_components].mean(axis=1)
df_diff[["name", "legendary_status", "difficulty_score"]].head()
```

```python
strength_med = df_diff["base_total"].median()
difficulty_med = df_diff["difficulty_score"].median()

def label_quadrant(row):
    strong = row["base_total"] >= strength_med
    hard = row["difficulty_score"] >= difficulty_med
    if strong and hard:
        return "Strong & Hard"
    if strong and not hard:
        return "Strong & Easy"
    if not strong and hard:
        return "Weak & Hard"
    return "Weak & Easy"
```

```python
df_diff["strength_difficulty_group"] = df_diff.apply(label_quadrant, axis=1)
df_diff["strength_difficulty_group"].value_counts()
```

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df_diff,
    x="base_total",
    y="difficulty_score",
    hue="strength_difficulty_group",
    style="legendary_status",
    alpha=0.7
)
plt.title("Strength vs Difficulty: Quadrant View")
plt.xlabel("Base Total (Strength)")
plt.ylabel("Difficulty Score (higher = harder)")
plt.tight_layout()
plt.savefig("../images/strength_vs_difficulty_quadrant_view.png")
plt.show()
```

```python
def top_n(group_name, n=10, sort_col="base_total", ascending=False):
    return (
        df_diff[df_diff["strength_difficulty_group"] == group_name]
        .sort_values(sort_col, ascending=ascending)
        [["name", "generation", "legendary_status", "base_total", "difficulty_score",
          "capture_rate", "base_egg_steps", "experience_growth"]]
        .head(n)
    )

top_n("Strong & Easy", n=10, sort_col="base_total", ascending=False)
```

## Let's look at the Pokemon types and their vulnerabilities now

```python
stat_colors = {
    "hp": "#FF9999",
    "defense": "#FFD700",
    "attack": "#FFB347",
    "sp_attack": "#9ACD32",
    "sp_defense": "#6495ED",
    "speed": "#DA70D6"
}
```

```python
def plot_all_types_radar_colored(dataframe, stat_cols, color_map, type_col="type1"):
    type_stat_avg = dataframe.groupby(type_col)[stat_cols].mean()
    types = type_stat_avg.index.tolist()

    n_cols = 4
    n_rows = int(np.ceil(len(types) / n_cols))

    angles = np.linspace(0, 2 * np.pi, len(stat_cols), endpoint=False).tolist()
    angles += angles[:1]

    fig, axs = plt.subplots(
        n_rows, n_cols,
        figsize=(n_cols * 4, n_rows * 4),
        subplot_kw=dict(polar=True)
    )
    axs = axs.flatten()

    for i, t in enumerate(types):
        values = type_stat_avg.loc[t].tolist()
        values += values[:1]

        dominant_stat = type_stat_avg.loc[t].idxmax()
        color = color_map.get(dominant_stat, "#AAAAAA")

        ax = axs[i]
        ax.plot(angles, values, linewidth=1.5, color=color)
        ax.fill(angles, values, alpha=0.25, color=color)
        ax.set_title(str(t), size=12, y=1.1)
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(stat_cols, size=8)
        ax.set_yticklabels([])

    for j in range(i + 1, len(axs)):
        fig.delaxes(axs[j])

    plt.suptitle(
        "Average Base Stats by Type (Radar Charts, Colored by Dominant Stat)",
        size=16,
        y=1.02
    )
    plt.tight_layout()
    plt.savefig("../images/average_stats_radar_charts.png")
    plt.show()

radar_stats = ["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]

plot_all_types_radar_colored(df, radar_stats, stat_colors, type_col="type1")
```
